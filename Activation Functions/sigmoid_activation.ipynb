## Sigmoid Activation Function

The Sigmoid activation function is a widely used activation function in machine learning and neural networks, particularly in the context of binary classification. It transforms any input value into a range between 0 and 1, effectively squashing the output. The mathematical representation of the Sigmoid function is:


$f(x) = \frac{1}{1 + e^{-x}}$


### Key Characteristics:
- **Output Range**: The output values range between 0 and 1, which can be interpreted as probabilities, making it suitable for binary classification tasks.
- **Non-linearity**: The Sigmoid function introduces non-linearity into the model, allowing it to learn complex patterns in the data.
- **Smooth Gradient**: It has a smooth gradient, which helps in gradient-based optimization techniques. However, it can suffer from vanishing gradient problems for very high or very low input values.



```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Input values 
x = np.linspace(-10, 10, 100)

# Apply Sigmoid to the input
sigmoid_output = sigmoid(x)

# Plot the graph 
plt.plot(x, sigmoid_output, label = 'Sigmoid', color = 'green')
plt.title('Sigmoid Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.legend()
plt.show()
```
