## ReLU (Rectified Linear Unit) Activation Function

The ReLU (Rectified Linear Unit) is one of the most widely used activation functions in neural networks, especially for deep learning models. It is defined as:


$f(x) = max(0, x)$


This means that the function returns 0 for any input value less than or equal to zero, and returns the input value itself for any positive input.

### Key Characteristics:
- **Non-linearity**: ReLU introduces non-linearity to the network, which helps model complex patterns in the data.
- **Simplicity**: Compared to other activation functions like sigmoid or tanh, ReLU is computationally efficient due to its simple max operation.
- **Sparsity**: ReLU often leads to sparsity in the neural network, as it outputs zero for negative inputs, making some neurons inactive, which can improve model efficiency.



```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def relu(x):
    return np.maximum(0, x)

# Input values
x = np.linspace(-10, 10, 100)

# Apply ReLU to the input
relu_output = relu(x)

# Plot the graph
plt.plot(x, relu_output, label='ReLU', color='blue')
plt.title('ReLU Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.legend()
plt.show()
```
